Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


Here are five plots you can create using the Seaborn library along with their uses:

1. **Histogram** (`sns.histplot`):
   - **Use**: Visualizes the distribution of a single continuous variable. Helps in understanding the frequency of data points within specified bins.

2. **Scatter Plot** (`sns.scatterplot`):
   - **Use**: Displays the relationship between two continuous variables. Useful for identifying correlations or patterns.

3. **Box Plot** (`sns.boxplot`):
   - **Use**: Shows the distribution of a continuous variable through quartiles and highlights outliers. Useful for comparing distributions across categories.

4. **Violin Plot** (`sns.violinplot`):
   - **Use**: Combines aspects of box plots and density plots to show the distribution of a continuous variable and its probability density. Useful for comparing distributions across categories with a richer view.

5. **Pair Plot** (`sns.pairplot`):
   - **Use**: Creates a matrix of scatter plots for each pair of variables in a dataset, along with histograms on the diagonal. Useful for exploring relationships and distributions among multiple variables.


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create the line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Show the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create the first box plot for 'age'
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')

# Create the second box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')

# Show the plots
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create the histogram for 'price' with hue for 'cut'
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')

# Show the plot
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create the pair plot with hue for 'species'
sns.pairplot(data=iris, hue='species')

# Show the plot
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the data to get the format needed for the heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Create the heatmap
sns.heatmap(data=flights_pivot, cmap="YlGnBu", annot=True, fmt="d")

# Show the plot
plt.title('Heatmap of Monthly Flight Passengers')
plt.show()
```
